```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = "Personlocal.xlsx"

col_BusinessEntityID = "BusinessEntityID"
col_PersonType = "PersonType"
col_NameStyle = "NameStyle"
col_Title = "Title"
col_FirstName = "FirstName"
col_MiddleName = "MiddleName"
col_LastName = "LastName"
col_Suffix = "Suffix"
col_EmailPromotion = "EmailPromotion"
col_ModifiedDate = "ModifiedDate"

df = pd.read_excel(excel_file)

# Count the number of NaN values in each column
#nan_count = df.isnull().sum()

count_Bei = df[col_BusinessEntityID].isna().sum()
count_Pt = df[col_PersonType].isna().sum()
count_Ns = df[col_NameStyle].isna().sum()
count_Tt = df[col_Title].isna().sum()
count_Fn = df[col_FirstName].isna().sum()
count_Mn = df[col_MiddleName].isna().sum()
count_Ln = df[col_LastName].isna().sum()
count_Su = df[col_Suffix].isna().sum()
count_Ep = df[col_EmailPromotion].isna().sum()
count_Md = df[col_ModifiedDate].isna().sum()


# Count the total no# of rows
number_of_rows = len(df)

# Difference with Nan data of each column
Diff_Bei = number_of_rows - count_Bei
Diff_Pt = number_of_rows - count_Pt
Diff_Ns = number_of_rows - count_Ns
Diff_Tt = number_of_rows - count_Tt
Diff_Fn = number_of_rows - count_Fn
Diff_Mn = number_of_rows - count_Mn
Diff_Ln = number_of_rows - count_Ln
Diff_Su = number_of_rows - count_Su
Diff_Ep = number_of_rows - count_Ep
Diff_Md = number_of_rows - count_Md

# Create data
categories = ["Bei","PT","NS","Tt","FN","MN","LN","Su","Ep","Md"]
values1 = [count_Bei,count_Pt,count_Ns,count_Tt,count_Fn,count_Mn,count_Ln,count_Su,count_Ep,count_Md]
values2 = [Diff_Bei,Diff_Pt,Diff_Ns,Diff_Tt,Diff_Fn,Diff_Mn,Diff_Ln,Diff_Su,Diff_Ep,Diff_Md]


# Plot the stacked bar chart
plt.bar(categories, values1,label="Count of Null data")
plt.bar(categories, values2, bottom=values1, label="Count of actual data")

# Add data labels
for i, value in enumerate(values1):
    plt.text(i, value, value, ha="right")
for i, value in enumerate(values2):
    plt.text(i, value + values1[i], value, ha="right")

# Increase the Graph size by extending the Y axis.
plt.ylim(0, 70)

    
plt.xlabel("Name of the columns")
plt.ylabel("Sum of Null and Actual Data in dataset")
plt.title("Null value population in experience study dataset")
plt.legend(loc='upper right')
plt.show()
```
