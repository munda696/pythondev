```python
import tkinter
from tkinter import *
from tkinter import messagebox,ttk
from tkinter import ttk
import xml.etree.ElementTree as ET

def update_xml():
 """Updates two tags in an XML file based on textbox/combobox entries and displays the updated content."""

 # Get text from textbox
 text1 = textbox1.get()

 # Get selection from combobox
 text2 = combobox2.get() # Access the selected value from the combobox

 # Parse the XML file (assuming the path is correct)
 try:
  tree = ET.parse(file_path) # Replace with your actual file path
  root = tree.getroot()
 except ET.ParseError as e:
  result_label.config(text="Error parsing XML: {}".format(e))
  return

 # Update tags based on user input
 try:
  root.find("StudyRun").text = text1
  root.find("SequenceStartDate").text = text2
 except (AttributeError, KeyError):
  result_label.config(text="Error: Tags 'StudyRun' or 'SequenceStartDate' not found!")
  return

 # Write the updated XML data to the file
 try:
  tree.write(file_path)
  result_label.config(text="XML Updated Successfully!")

  # Get updated XML content as a string
  updated_content = ET.tostring(root, encoding='utf-8').decode('utf-8')

  # Clear the text area and display the updated content
  text_area.delete("1.0", tk.END)
  text_area.insert(tk.INSERT, updated_content)
 except OSError as e:
  result_label.config(text="Error writing to file: {}".format(e))  
   
root = tkinter.Tk()
    
# Geometry Definitions
root.title("Measure - Experience study processing")
root.geometry('700x500')

# Defining Labels
lbl_Header = tkinter.Label(root, text="Experience study processing", font=('Comic Sans MS', 15, 'bold')).grid(row=0, column=0, columnspan=2)  # Span across two columns

# Defining AutomationSequence fields
lbl_confg = tkinter.Label(root, text="Select Measure automation sequence", font=('Comic Sans MS', 10, 'bold')).grid(row=2, column=0)
# List of Automation Sequences
choices = ['ULH_ProductionSequenceAutomation', 'AH_ProductionSequenceAutomation', 'Income_Pay', 'SPIA', 'Mercer']
Combo_Esp = ttk.Combobox(root, values=choices, width=35).grid(row=2, column=1)  # Grid placement adjusted


#lbl_spacing1 = tkinter.Label(root,text="-",font=('Comic Sans MS',2)).grid(row=3,column=0)
#lbl_spacing2 = tkinter.Label(root,text="-",font=('Comic Sans MS',2)).grid(row=4,column=0)

#Config labels
lbl_cofg1 = tkinter.Label(root, text="Select fields to update config file:", font=('Comic Sans MS', 10, 'bold')).grid(row=5, column=0)

lbl_SR = tkinter.Label(root, text="<Study Run>").grid(row=7, column=0)
textboxSR = tkinter.Entry(root, width=30).grid(row=7, column=1)

lbl_SSD = tkinter.Label(root, text="<Sequence start date>").grid(row=8, column=0)
textboxSSD = tkinter.Entry(root, width=30).grid(row=8, column=1)

# Combobox for selecting true/false
combobox2_label = tkinter.Label(root, text="SequenceStartDate Value:")
combobox2_values = ("true", "false")  # List of values for the combobox
combobox2 = ttk.Combobox(root, values=combobox2_values, state="readonly")  # Readonly combobox
combobox2.set(combobox2_values[0])  # Set default selection to "true"
combobox2.grid(row=8, column=2)  # Grid placement adjusted

#Config textboxes


#Submit button
#lbl_spacing2 = tkinter.Label(root,text="-",font=('Comic Sans MS',2)).grid(row=13,column=0)
#Submit_but = tkinter.Button(root,text="Submit",command = OnClick_Submit,width=15).grid(row=14,column=1)
#Submit_but = tkinter.Button(root,text="Submit",width=15).grid(row=14,column=1)

# Outout box to display logs
#Output = Text(root, height = 8,width = 45,bg = "light cyan").grid(row=17,column=1)


# Button to trigger update
update_button = tkinter.Button(root, text="Update XML", command=update_xml).grid(row=13, column=1)  # Grid placement adjusted

# Label to display results
result_label = tkinter.Label(root, text="")
result_label.grid(row=14, column=0, columnspan=2)  # Span across two columns

# Text area to display initial and updated XML content
text_area = tkinter.Text(root)
#text_area.pack(expand=True, fill=tk.BOTH)  # Uncomment to allow expansion

# Start the main event loop
root.mainloop()
```
